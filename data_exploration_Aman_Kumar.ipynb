
# Load the Data

```python
import pandas as pd

# Load the dataset
file_path = 'Global-Superstore - Global-Superstore.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

# Data Exploration

```python
# Check the shape of the dataset
data.shape
```

```python
# Get basic information about the dataset
data.info()
```

```python

# Summary statistics for numerical columns
data.describe()
```

```python
# Check for missing values
data.isnull().sum()
```

Here, we can see that there is only missing value in Postal code

```python
# Display rows with missing Postal Codes
missing_postal_code_data = data[data['Postal Code'].isnull()]
missing_postal_code_data.head()
```

```python
# Display rows with given Postal Codes
postal_code_data = data[data['Postal Code'].notnull()]
postal_code_data.head()
```

```python
# Count of missing postal codes by country
missing_postal_by_country = missing_postal_code_data['Country'].value_counts()
print(missing_postal_by_country)
```

```python

# Count of non-missing postal codes by country
non_missing_postal_code_data = data.dropna(subset=['Postal Code'])
non_missing_postal_by_country = non_missing_postal_code_data['Country'].value_counts()
print(non_missing_postal_by_country)
```

The majority of missing postal codes are associated with countries such as Australia (2,837 missing), France (2,827 missing), and Mexico (2,644 missing).
After filtering out the missing postal codes, we found that the remaining postal codes are primarily associated with the United States.
The lack of postal codes for international transactions might impact the completeness and accuracy of geographic analysis outside the United States.

```python
# Verify if non-missing postal codes are only for United States
unique_countries_with_postal = non_missing_postal_code_data['Country'].unique()
print(unique_countries_with_postal)
```

Droping Postal Code column

```python
# Drop the 'Postal Code' column
data = data.drop(columns=['Postal Code'])

# Verify the column has been removed
print(data.columns)
```

# Data Visualization


Calculate and Visualize Top 10 Total Sales by country

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate total sales by country
total_sales_by_country = data.groupby('Country')['Sales'].sum().reset_index()

# Get top 10 countries by sales
top_10_countries_by_sales = total_sales_by_country.sort_values(by='Sales', ascending=False).head(10)

# Visualize top 10 countries by sales
plt.figure(figsize=(14, 8))
sns.barplot(x='Sales', y='Country', data=top_10_countries_by_sales, palette='magma')
plt.title('Top 10 Countries by Sales')
plt.xlabel('Sales ($)')
plt.ylabel('Country')
plt.show()
```

Showing Top 10 Countries by sales which indicate that US has most sales which mean more profit and high quantity of Import and export


# Sales by Product Category

```python
# Calculate sales by product category
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()

# Visualize sales by product category
plt.figure(figsize=(10, 6))
sns.barplot(x='Sales', y='Category', data=sales_by_category, palette='viridis')
plt.title('Sales by Product Category')
plt.xlabel('Sales ($)')
plt.ylabel('Category')
plt.show()
```

Technology sold the most, about 5 million dollars. Furniture sold next, around 4 million dollars. Office Supplies sold the least, which is sightly less than Furniture, around 3.7 million dollars.


Top Selling Products

```python
# Calculate sales by product
sales_by_product = data.groupby('Product Name')['Sales'].sum().reset_index()

# Get top 10 selling products
top_selling_products = sales_by_product.sort_values(by='Sales', ascending=False).head(10)

# Visualize top selling products
plt.figure(figsize=(12, 8))
sns.barplot(x='Sales', y='Product Name', data=top_selling_products, palette='magma')
plt.title('Top 10 Selling Products')
plt.xlabel('Sales ($)')
plt.ylabel('Product Name')
plt.show()
```

Top 10 Selling Products," displays the sales of various products. The Apple Smart Phone, Full Size, tops the list with about 80,000 units sold, followed closely by the Cisco Smart Phone and more

Sales Trend Over Time

```python
# Convert Order Date to datetime
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Calculate monthly sales
monthly_sales = data.set_index('Order Date').resample('M')['Sales'].sum().reset_index()

# Visualize sales trend over time
plt.figure(figsize=(14, 8))
sns.lineplot(x='Order Date', y='Sales', data=monthly_sales, marker='o', color='b')
plt.title('Sales Trend Over Time')
plt.xlabel('Order Date')
plt.ylabel('Sales ($)')
plt.show()
```

Monthly Sales Comparison

```python
# Extract year and month from Order Date
data['Year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.month

# Calculate monthly sales for each year
monthly_sales_comparison = data.groupby(['Year', 'Month'])['Sales'].sum().unstack(level=0).fillna(0)

# Visualize monthly sales comparison
plt.figure(figsize=(14, 8))
monthly_sales_comparison.plot(kind='line', marker='o')
plt.title('Monthly Sales Comparison')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.legend(title='Year')
plt.show()
```

# Multivariate Analysis


Sales by Category and Sub-Category

```python
# Bar plot of Sales by Category and Sub-Category
plt.figure(figsize=(14, 10))
category_subcategory_sales = data.groupby(['Category', 'Sub-Category'])['Sales'].sum().reset_index()
sns.barplot(x='Sales', y='Sub-Category', hue='Category', data=category_subcategory_sales)
plt.title('Sales by Category and Sub-Category')
plt.xlabel('Sales')
plt.ylabel('Sub-Category')
plt.legend(title='Category')
plt.show()
```

The bar chart titled "Sales by Category and Sub-Category" displays sales data across different categories and their sub-categories.

Sales by Region and Category

```python
# Bar plot of Sales by Region and Category
plt.figure(figsize=(14, 8))
region_category_sales = data.groupby(['Region', 'Category'])['Sales'].sum().reset_index()
sns.barplot(x='Sales', y='Region', hue='Category', data=region_category_sales)
plt.title('Sales by Region and Category')
plt.xlabel('Sales')
plt.ylabel('Region')
plt.legend(title='Category')
plt.show()
```

# Time Series Analysis

 Sales Trend Over Time

```python
# Convert 'Order Date' to datetime format
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Extract year and month from 'Order Date'
data['Year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.to_period('M')
```

Calculate Monthly Sales

```python
# Monthly Sales Trend
monthly_sales = data.groupby('Month')['Sales'].sum().reset_index()

# Convert 'Month' back to datetime format
monthly_sales['Month'] = monthly_sales['Month'].dt.to_timestamp()
```

```python

# Line plot of Monthly Sales Trend
plt.figure(figsize=(14, 8))
sns.lineplot(x='Month', y='Sales', data=monthly_sales)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

The line graph titled "Monthly Sales Trend" illustrates the sales figures over time from January 2011 to April 2015.

At first, from early 2011 to mid-2012, sales were fairly steady with a few small ups and downs. However, starting in late 2012, there was a big increase in sales, which continued to rise over the next few years.
